```python
import numpy as np

def sigmoid(x):
    return(1/(1+np.exp(-x)))
    
# Input values
x1 = 1
x2 = 2
x3 = 3

# Input vector
x = np.array([[x1],[x2],[x3]])

# Layer 1 : (x -> [-2, 4])

W1 = np.array([[ 0.8,  0.1, -0.6], [-0.5, -0.4,  1.2]])

b1 = np.array([[-2], [ 4]])

z1 = np.dot(W1, x) + b1
h1 = sigmoid(z1)

print("Layer 1 Output")
print(h1)


# Layer 2 : ([-2,4] -> [-2.5,1.5,-0.8,1.6])

W2 = np.array([[-1.2,  1.4],[ 0.6,  0.1],[-0.2,  0.5],[ 2.0, -1.5]])
b2 = np.array([[-2.5],[ 1.5],[-0.8],[ 1.6]])

z2 = np.dot(W2, h1) + b2
h2 = sigmoid(z2)

print("Layer 2 Output")
print(h2)


# Layer 3 : ([-2.5,1.5,-0.8,1.6] -> [-1.4,2.2])

W3 = np.array([[ 1.6,  0.1,  0.4, -0.4],[-1.0, -1.2,  0.8, -1.4]])

b3 = np.array([[-1.4],[ 2.2]])

z3 = np.dot(W3, h2) + b3
h3 = sigmoid(z3)

print("Layer 3 Output")
print(h3)


# Output Layer : ([-1.4,2.2] -> [-1,-2,-3])

W4 = np.array([[ 0.5,  0.8],[-1.2, -1.0],[-0.6, -0.2]])

b4 = np.array([[-1],[-2],[-3]])

z4 = np.dot(W4, h3) + b4
y = sigmoid(z4)

print("Final Output")
print(y)
```

```python
#10d)
import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))

class layer():

    def __init__(self, n_inputs, n_neurons):
        self.n_inputs = n_inputs
        self.n_neurons = n_neurons
        self.weights = np.zeros((n_neurons, n_inputs))
        self.biases = np.zeros(n_neurons)

    def activations(self, x):
        return sigmoid(np.dot(self.weights, x) + self.biases)
   
```

```python
#(i)
lay=layer(3,1)
lay.weights=np.array([-0.8,0.5,1])
lay.biases=2
x=np.array([1,-2,-1])
lay.activations(x)
print("(i)")
print(x,y)

#(ii)
lay1=layer(3,2)
lay1.weights=np.array([[0.8,-1,-0.4],[0.2,-2,-0.5]])
lay1.biases=np.array([2,1])
lay2=layer(2,1)
lay2.weights=np.array([[2,3]])
lay2.biases=-1
lay_list=[lay1,lay2]
x=np.array([0.4,-2.8,1.2])
a=x
for lay in lay_list:
    a=lay.activations(a)
y = a

print("(ii)")
print(x,y)



# (iii)
lay1 = layer(2,2)
lay1.weights = np.array([[ 1, -1],[-1,  1]])
lay1.biases = np.array([-1.2, -1.2])


lay2 = layer(2,1)
lay2.weights = np.array([[1,1]])
lay2.biases = np.array([-0.5])

# Input
x = np.array([1,-2])

a1 = lay1.activations(x)
y = lay2.activations(a1)

print("(iii)")
print(x,y)
```

```python
#11-1
import numpy as np

def grad_descent(x_start,f,f_prime,eta=0.1,iter=10):
    
    x=x_start
    for j in range(iter):
        x_star=x-eta*f_prime(x) #iteration step
        print(f"{j+1} {np.round(x,2)} {np.round(f_prime(x),2)} {np.round(x_star,2)} {np.round(f(x_star),2)}")
        x=x_star
        
```
