```python
import math

def q_solve(a, b, c):
    if a == 0:
        return "Not a quadratic equation"
    
    D = b**2 - 4*a*c
    
    if D > 0:
        x1 = (-b + math.sqrt(D)) / (2*a)
        x2 = (-b - math.sqrt(D)) / (2*a)
        return x1, x2
    elif D == 0:
        x = -b / (2*a)
        return x
    else:
        return "no real solution"
```

```python
print(q_solve(1, -3, 2))   # (2.0, 1.0)
print(q_solve(1, 2, 1))    # -1.0
print(q_solve(1, 0, 1))    # no real solution
```
