```python
#full code
# Import symbols function from SymPy
from sympy import symbols

# Define a Python function using
# logical operators: and, or, not
def l_term(X, Y, Z):
    return (X and not Y) or (not (Z or Y) or X)

# Evaluate the function for the given inputs
print(l_term(True,True,False))
print(l_term(False,True,False))
print(l_term(False,False,True))

#Define the logical term in SymPy
X, Y, Z = symbols('X Y Z')

expr = (X & ~Y) | (~(Z | Y) | X)

#Evaluate using True and False values
print(expr.subs({X:True, Y:True, Z:False}))
print(expr.subs({X:False, Y:True, Z:False}))
print(expr.subs({X:False, Y:False, Z:True}))

# Evaluate again using 1 and 0
# (1 = True, 0 = False)
print(expr.subs({X:1, Y:1, Z:0}))
print(expr.subs({X:0, Y:1, Z:0}))
print(expr.subs({X:0, Y:0, Z:1}))
```

```python
from sympy import *
from sympy.logic.boolalg import truth_table, simplify_logic
```

```python
#i) (A ∨¬B)∧¬(¬A∧B)
A, B = symbols('A B')

expr = (A | ~B) & ~(~A & B)

print("Original:", expr)

for row in truth_table(expr, [A, B]):
    print(row)

simp = simplify_logic(expr)

print("Simplified:", simp)

for row in truth_table(simp, [A, B]):
    print(row)
```

```python
#ii) (¬x1 → x2)∨(x1 ∧¬x2)
x1, x2 = symbols('x1 x2')

expr = ((~~x1) | x2) | (x1 & ~x2)

print("Original:", expr)

for row in truth_table(expr, [x1, x2]):
    print(row)

simp = simplify_logic(expr)

print("Simplified:", simp)

for row in truth_table(simp, [x1, x2]):
    print(row)
```

```python
#iii) (X ∧ ¬Y ∧Z)⊕(¬X ∧Y)⊕(Y ∧¬Z)
X, Y, Z = symbols('X Y Z')

expr = (X & ~Y & Z) ^ (~X & Y) ^ (Y & ~Z)

print("Original:", expr)

for row in truth_table(expr, [X, Y, Z]):
    print(row)

simp = simplify_logic(expr)

print("Simplified:", simp)

for row in truth_table(simp, [X, Y, Z]):
    print(row)
```

```python
# iv) (u∧v ↔v∧w)→(u⊕¬w)
u, v, w = symbols('u v w')

expr = ~Equivalent(u & v, v & w) | (u ^ ~w)

print("Original:", expr)

for row in truth_table(expr, [u, v, w]):
    print(row)

simp = simplify_logic(expr)

print("Simplified:", simp)

for row in truth_table(simp, [u, v, w]):
    print(row)
```

```python
#v) (¬(a ∨¬b)⊕(¬c∨d)) → (¬a∧b∧¬c)
a, b, c, d = symbols('a b c d')

expr = ~(~(a | ~b) ^ (~c | d)) | (~a & b & ~c)

print("Original:", expr)

for row in truth_table(expr, [a, b, c, d]):
    print(row)

simp = simplify_logic(expr)

print("Simplified:", simp)

for row in truth_table(simp, [a, b, c, d]):
    print(row)
```

```python
#vi) (X1 ∧X2) ↔ (¬X3∨X4)
X1, X2, X3, X4 = symbols('X1 X2 X3 X4')

expr = Equivalent(X1 & X2, ~X3 | X4)

print("Original:", expr)

for row in truth_table(expr, [X1, X2, X3, X4]):
    print(row)

simp = simplify_logic(expr)

print("Simplified:", simp)

for row in truth_table(simp, [X1, X2, X3, X4]):
    print(row)
```

```python
# 3(C)
from sympy import *
from sympy.logic.inference import satisfiable

# Variables
A, B = symbols('A B')
x1, x2 = symbols('x1 x2')
X, Y, Z = symbols('X Y Z')
u, v, w = symbols('u v w')
a, b, c, d = symbols('a b c d')
X1, X2, X3, X4 = symbols('X1 X2 X3 X4')

# i
expr1 = (A | ~B) & ~(~A & B)

# ii
expr2 = ((~~x1) | x2) | (x1 & ~x2)

# iii
expr3 = (X & ~Y & Z) ^ (~X & Y) ^ (Y & ~Z)

# iv
expr4 = ~Equivalent(u & v, v & w) | (u ^ ~w)

# v
expr5 = ~(~(a | ~b) ^ (~c | d)) | (~a & b & ~c)

# vi
expr6 = Equivalent(X1 & X2, ~X3 | X4)

expressions = [expr1, expr2, expr3, expr4, expr5, expr6]

for i, expr in enumerate(expressions, start=1):

    print(f"\n===== Expression {i} =====")

    print("One satisfying model:")
    print(satisfiable(expr))

    print("\nAll satisfying models:")
    for model in satisfiable(expr, all_models=True):
        print(model)
```
